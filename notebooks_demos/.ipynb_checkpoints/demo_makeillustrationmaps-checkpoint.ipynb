```python
import os
import glob
import sys
sys.path.append('..')
import numpy as np 
import xarray as xr 
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
GS_info = {
    'name' : 'GS'                                # region name
    'lon_min' : 309.,                            # domain min longitude
    'lon_max' : 314.,                            # domain max longitude
    'lat_min' : 30.,                             # domain min latitude
    'lat_max' : 42.                              # domain max latitude
}

Med_info = {
    'name' : 'Med'                               # region name
    'lon_min' : 1.,                              # domain min longitude
    'lon_max' : 20.,                             # domain max longitude
    'lat_min' : 30.,                             # domain min latitude
    'lat_max' : 45.                              # domain max latitude
}


ACC_info = {
    'name' : 'ACC'                              # region name
    'lon_min' : 1.,                             # domain min longitude
    'lon_max' : 2.,                             # domain max longitude
    'lat_min' : 1.,                             # domain min latitude
    'lat_max' : 2.                              # domain max latitude
}
```

# A. 21 day orbit

## Reading 21 day SWOT data

```python
# Data repository
path_data = '../swot_data/swot_21j_share' 

ds_pass = xr.open_mfdataset(path_data+'/SWOT*',combine='nested',concat_dim='num_lines')
```

## Selecting and saving Gulf Stream region

```python
%%time

region_info = GS_info

ds_pass1 = ds_pass.where(ds_pass['longitude']<region_info['lon_max'],drop=True)
ds_pass1 = ds_pass1.where(ds_pass1['longitude']>region_info['lon_min'],drop=True)
ds_pass1 = ds_pass1.where(ds_pass1['latitude']<region_info['lat_max'],drop=True)
ds_pass1 = ds_pass1.where(ds_pass1['latitude']>region_info['lat_min'],drop=True)

ds_pass1.to_netcdf(path_data+'_'+region_info['name']+'.nc')

del ds_pass1
```

## Selecting and saving Mediterranean region

```python
%%time

region_info = Med_info

ds_pass1 = ds_pass.where(ds_pass['longitude']<region_info['lon_max'],drop=True)
ds_pass1 = ds_pass1.where(ds_pass1['longitude']>region_info['lon_min'],drop=True)
ds_pass1 = ds_pass1.where(ds_pass1['latitude']<region_info['lat_max'],drop=True)
ds_pass1 = ds_pass1.where(ds_pass1['latitude']>region_info['lat_min'],drop=True)

ds_pass1.to_netcdf(path_data+'_'+region_info['name']+'.nc')

del ds_pass1
```

## Selecting and saving Antarctic Circumpolar region

```python
%%time

region_info = ACC_info

ds_pass1 = ds_pass.where(ds_pass['longitude']<region_info['lon_max'],drop=True)
ds_pass1 = ds_pass1.where(ds_pass1['longitude']>region_info['lon_min'],drop=True)
ds_pass1 = ds_pass1.where(ds_pass1['latitude']<region_info['lat_max'],drop=True)
ds_pass1 = ds_pass1.where(ds_pass1['latitude']>region_info['lat_min'],drop=True)

ds_pass1.to_netcdf(path_data+'_'+region_info['name']+'.nc')

del ds_pass1
```

```python
%%time 

# Load a shapefile containing continent boundaries
import cartopy.crs as ccrs
import cartopy.feature as feature

import matplotlib.patches as patches

fig, ax = plt.subplots(subplot_kw={'projection': ccrs.PlateCarree()}, figsize=(10, 6))
plt.title('1 day orbit',fontsize=22)


plt.scatter(ds_pass.longitude[:,10:12],ds_pass.latitude[:,10:12],s=1,c=ds_pass.ssha_denoised_unet[:,10:12])

ax.set_xlim(-180, 180)
ax.set_ylim(-90, 90)
ax.set_xticks(range(-180, 181, 60))
ax.set_xticklabels(['180$^\circ$ W','120$^\circ$ W','60$^\circ$ W', '0$^\circ $','60$^\circ$ E','120$^\circ$ E','180$^\circ$ E'],fontsize=12)
ax.set_yticks(range(-90, 91, 30))
ax.set_yticklabels(['90$^\circ$ S','60$^\circ$ S','30$^\circ$ S', '0$^\circ $','30$^\circ$ N','60$^\circ$ N','90$^\circ$ N'],fontsize=12)
ax.set_xlabel('Longitude',fontsize=16)
ax.set_ylabel('Latitude',fontsize=16)


# Add continents with a different color
ax.add_feature(feature.LAND) 

rect = patches.Rectangle((-51, 30), 5, 12, linewidth=2, edgecolor='r', facecolor='none')
ax.add_patch(rect)

rect = patches.Rectangle((0, 35), 5, 8, linewidth=2, edgecolor='r', facecolor='none')
ax.add_patch(rect)

rect = patches.Rectangle((5, -60), 8, 18, linewidth=2, edgecolor='r', facecolor='none')
ax.add_patch(rect)


plt.savefig('../figures/illustration_1daySWOTorbit.png')

# Show the plot
plt.show()
```
