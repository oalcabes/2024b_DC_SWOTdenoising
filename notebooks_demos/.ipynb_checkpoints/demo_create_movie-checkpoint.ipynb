```python
path_data = '../swot_data/'
path_denoised = '../denoised_data/' 
method_name = 'gomezfilt'
 
region = 'GS'    
```

```python
all_files = sorted(glob.glob(path_data+'*'+region+'*'))
```

```python
ds_passes = xr.open_mfdataset(all_files,combine='nested',concat_dim='num_lines')
```

```python
create_movie(ds_passes, name_var = 'ssha_unedited', method='SWOT', region = region, lonlat_minmax=[lon_min,lon_max,lat_min,lat_max], time0 = np.datetime64('2023-06-03'),
                 ndays = 38, dir_output='../figures/', dim_name=['time','latitude','longitude'], 
                 framerate=3, Display=True, clim=[-0.6,0.6], cmap='Spectral', newmovie=True)
```

```python
grad_ssha_noiseless, lap_ssha_noiseless = deriv1and2(ds_passes.ssha_noiseless)
```
